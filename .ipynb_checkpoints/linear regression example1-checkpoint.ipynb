```python
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
import PIL.Image as pilimg
import plotly.offline as pyo
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
x = [1, 2, 3]
y = [1, 2, 3]
```

```python
# Variable 함수도 placeholder와 비슷하게 데이터를 담는 그릇이다.
# 아래 그림과 같은 초기값들이 존재한다.
w = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
variable = pilimg.open('C:\\Users\\Msi\\Desktop\\딥러닝\\variable 초기값.png')
variable
```

```python
# 가설 설정(= 추정 회귀식 = 예측 값 = predict)
hypothesis = x * w + b
```

```python
# cost를 최저로 하기 위한 최소제곱법(mse) 설정
cost = tf.reduce_mean(tf.square(hypothesis - y))
```

```python
# learning_rate를 0.01로 두고 경사하강법 설정
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
```

```python
# 최소화하는 과정
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
step_list = []
cost_list = []
```

```python
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(w), sess.run(b))
        step_list.append(step)
        cost_list.append(sess.run(cost))
```

```python
# 데이터가 별로 없기 때문에 급격히 cost값이 0으로 하는 것을 볼 수 있다.
sns.lineplot(step_list, cost_list)
plt.xlabel('step', fontsize = 13)
plt.ylabel('cost', fontsize = 13)
plt.show()
```
