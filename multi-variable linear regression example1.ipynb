```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x1 = [73, 93, 89, 96, 73]
x2 = [80, 88, 91, 98, 66]
x3 = [75, 93, 30, 100, 70]
y = [152, 185, 180, 196, 142]

w1 = tf.Variable(tf.random_normal([1]), name = 'w1')
w2 = tf.Variable(tf.random_normal([1]), name = 'w2')
w3 = tf.Variable(tf.random_normal([1]), name = 'w3')
b = tf.Variable(tf.random_normal([1]), name = 'b')

hypothesis = x1*w1 + x2*w2 + x3*w3 + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.00001)
train = optimizer.minimize(cost)
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
# 처음에 learning_rate를 0.01로 잡았으나 cost값이 inf로 나와 0.001로 줄였다.
# 그래도 cost값이 튀어나가 0.0001 그 다음 0.00001로 줄이니 결과가 나왔다.
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train])
    if step % 10 == 0:
        print(step, 'cost : ', cost_val, '\npredict : ', hy_val)
```
